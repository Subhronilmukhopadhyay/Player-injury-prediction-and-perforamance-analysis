```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
import tensorflow as tf
```

```python
dataset = pd.read_csv("merged.csv")

list1 = []
for i in range(len(dataset)):
    list1.append(0)
    if(dataset['DAYS MISSED'][i] >0):
        list1[i]=1
    else:
        list1[i]=0
dataset['injured check'] = list1
```

```python
dataset['INJURED_TYPE'] = dataset['INJURED_TYPE'].fillna('No_injury')
print(dataset['INJURED_TYPE'])
```

```python
X = dataset.iloc[:, 3:-6].values
X = X.tolist()
for i in range(len(X)):
    for j in range(len(X[i])):
        if(X[i][j] ==' None'):
            X[i][j] = 0
        if(type(X[i][j]) == str):
            X[i][j] = float(X[i][j])
X = np.array(X)
y = dataset.iloc[:, -2].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print("Class Mapping:", class_mapping)
L = list(class_mapping.keys())
print(L)
unique_values_after_encoding = set(dataset['INJURED_TYPE'])
print("Unique Values in INJURED_TYPE After Label Encoding:", unique_values_after_encoding)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = tf.keras.models.Sequential()
```

```python

ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
```

```python
num_classes = len(np.unique(y_train))
ann.add(tf.keras.layers.Dense(units=num_classes, activation='softmax'))
```

```python
ann.compile(optimizer= 'adam', loss= 'sparse_categorical_crossentropy', metrics= ['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size= 150, epochs= 400 , validation_data=(X_test, y_test))
```
